## *MATPLOT LIB



```python
#iris dataset
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#working with default dataset

data = sns.load_dataset('iris')

data.head()
```

```python
#Matplotlib

#line graph

plt.plot(data['sepal_width'])
plt.show()
```

```python
plt.plot(data['petal_length'])
plt.title("Line chart using Petal_Length")
plt.xlabel("Indexes")
plt.ylabel("Petal Length")
plt.show()
```

```python
plt.plot(data['sepal_length'], color = 'black')  #plot() method to plot any numerical data
plt.title("Line chart for Sepal_length", fontsize=22)
plt.xlabel("Indexes")
plt.ylabel("Sepal Length")
plt.show()
```

```python
#incorportating multiple parameters in line chart
```

```python
plt.figure(figsize=(12,8))

plt.plot(data['petal_width'], color = 'black', label = "petal_width")
plt.plot(data['sepal_width'], color = 'green', label = "sepal_width")
plt.plot(data['petal_length'], color = 'orange', label = "petal_length")
plt.plot(data['sepal_length'], color = 'red', label = "sepal_length")
plt.xlabel("index")
plt.ylabel("Multiple features")
plt.title("Iris Species", fontsize=23)
plt.legend()
plt.show()
```

```python
#Adding dashed lines

plt.figure(figsize=(12,8))

plt.plot(data['petal_width'], color = 'black', label = "petal_width", linestyle="--")
plt.plot(data['sepal_width'], color = 'green', label = "sepal_width", marker='*')
plt.plot(data['petal_length'], color = 'orange', label = "petal_length", linestyle="--")
plt.plot(data['sepal_length'], color = 'red', label = "sepal_length", marker = '+')
plt.xlabel("index")
plt.ylabel("Multiple features")
plt.title("Iris Species", fontsize=23)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot(data['petal_width'], color = 'black', label = "petal_width", linestyle="--")
plt.plot(data['sepal_width'], color = 'green', label = "sepal_width", marker='*', linewidth=3)
plt.plot(data['petal_length'], color = 'orange', label = "petal_length", linestyle="--")
plt.plot(data['sepal_length'], color = 'red', label = "sepal_length", marker = '+', linewidth=3)
plt.xlabel("index")
plt.ylabel("Multiple features")
plt.title("Iris Species", fontsize=23)
plt.legend()
plt.show()
```

```python
#Adding grid lines

plt.figure(figsize=(12,8))
plt.grid()

plt.plot(data['petal_width'], color = 'black', label = "petal_width", linestyle="--")
plt.plot(data['sepal_width'], color = 'green', label = "sepal_width", marker='*', linewidth=3)
plt.plot(data['petal_length'], color = 'orange', label = "petal_length", linestyle="--")
plt.plot(data['sepal_length'], color = 'red', label = "sepal_length", marker = '+', linewidth=3)
plt.xlabel("index")
plt.ylabel("Multiple features")
plt.title("Iris Species", fontsize=23)
plt.legend()
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#histogram

plt.hist(tips['total_bill'], bins = 20, edgecolor = 'black')
plt.show()
```

```python
plt.pie(tips['total_bill'])
plt.show()
```

```python
day = tips.groupby('day')['total_bil'].sum().reset_index()
x=day['day']
y=day['total_bill']
explode = [0,0,0,0.1]
plt.pie(y, labels=x, explode = explode, autopct='%1.1f%%', colors = ['red', 'blue', 'green', 'yellow'], wedgeprops={'edgecolor: "black"})
```

```python
plt.boxplot()
```

```python
#Scatterplot
plt.scatter(tips['total_bill'], tips['tip'])
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel('Tip')
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], c = 'red')
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel('Tip')
plt.show()
```

```python
#color bar

plt.scatter(tips['total_bill'], tips['tip'] , s=100, c = tips['size'])
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel('Tip')
cbar=plt.colorbar()
cbar.set_label("Size")
plt.show()
```
